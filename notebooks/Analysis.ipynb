```python
new_cons_df=constructor_standings_df.merge(new_df,on=['raceId','constructorId'], how='left')
```

```python
#look up the rows with win==1
cons_no1=new_cons_df.loc[new_cons_df['wins']==1]
cons_no1_new=cons_no1[['wins','name']]
```

```python
#group the winners by their names and take the count of wins within each constructor
cons_gb=cons_no1_new.groupby(['name']).count()
```
