## Races and circuits

```python
races_df=pd.read_csv("../data/raw/races.csv")
circuits_df=pd.read_csv("../data/raw/circuits.csv")
```

```python
races_df.rename(columns={'name':'race_name','url':'races_url'},inplace=True)
circuits_df.rename(columns={'name':'circuit_name','url':'circuits_url'},inplace=True)
```

```python
races_circuits_df=races_df.merge(circuits_df,on='circuitId',how="inner")
```

```python
races_circuits_df.drop(columns=['alt'],inplace=True)
```

```python
races_circuits_df
```

```python
# I can see time will be a problem, I will set all the times to seconds for all dataframes with time intervals: 
def hh_mm_ss2seconds(hh_mm_ss):
    return reduce(lambda acc, x: acc*60 + x, map(float, hh_mm_ss.split(':')))
```

```python
races_circuits = races_circuits_df.replace({'time': '\\N'}, 0)
races_circuits
```

```python
races_circuits.to_csv('../data/Modified/races_circuits.csv', index = False,header=True)
```

```python
races_circuits_df1 = pd.read_csv('../data/Modified/races_circuits.csv',
                     converters={'time': hh_mm_ss2seconds})
```
