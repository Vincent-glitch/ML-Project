```python
import pandas as pd
```

```python
drivers_df = pd.read_csv("drivers.csv")
laps_df = pd.read_csv("lap_times.csv")
```

```python
Drivers_merge = drivers_df.merge(laps_df, left_on='driverId', right_on='driverId')
Drivers_merge
```

```python
constructors_df = pd.read_csv("constructors.csv")
constructors_df 
```

```python
constructors_standings = pd.read_csv("constructor_standings.csv")
constructors_standings
```

```python
Constructor_merge = constructors_df.merge(constructors_standings, left_on='constructorId', right_on='constructorId')
Constructor_merge
```

```python
Races_df = pd.read_csv("races.csv")
```

```python
Constructor_Final = Constructor_merge.merge(Races_df, left_on='raceId', right_on='raceId')
Constructor_Final.rename(columns={"name_x":"Constructor_Names","name_y":"Race_Name"},inplace=True)
```

```python
Constructor_Final.drop(columns=["url_x","url_y"],inplace=True)
```

```python
Constructor_Final
```

```python
Constructor_Final.to_csv("../Modified/constructor_final.csv")
```

```python
driver_standings = pd.read_csv("driver_standings.csv")
```

```python
Drivers_merge = drivers_df.merge(driver_standings, left_on='driverId', right_on='driverId')
Drivers_merge 
```

```python
Drivers_New = Drivers_merge.merge(Races_df, left_on='raceId', right_on='raceId')
Drivers_New
```

```python
Drivers_New.drop(columns=["url_x","url_y"],inplace=True)
```

```python
Drivers_New.to_csv("../Modified/Driver_Info.csv",header=True,index=False)
```
